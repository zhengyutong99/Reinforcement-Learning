```python
import pandas as pd
import matplotlib.pyplot as plt
import re

with open('/home/zhengyutong/data/Reinforce_Learning/Lab1/Training_process/6tuples_demo_training.txt', 'r') as f:
    episodes = f.readlines()
    episodes = [str(e.strip()) for e in episodes]

mean_values = []
for line in episodes:
    print(line)
    # match = re.search(r'mean = (\d+\.\d+)', line)
    match = re.search(r'mean = (\d+(?:\.\d+)?)', line)
    if match:
        mean_values.append(float(match.group(1)))

print(mean_values)
print(len(mean_values))
```

```python
x_values = [i*1000 for i in range(len(mean_values))]
# x_ticks = [j for j in range(0,1000001,200000)]
print(x_values)
# print(x_ticks)

plt.plot(x_values, mean_values, label = "Mean")
plt.ticklabel_format(style='plain', axis='x')
plt.legend()
plt.xlabel("Episodes")
plt.ylabel('Mean')
plt.title('Mean Values of 1000,000 Episodes')
# plt.xticks()
# plt.xticks(x_ticks)
plt.figure(figsize=(20, 20))
plt.show()
```

```python
x_values = [i*1000 for i in range(200)]
plt.plot(x_values, mean_values[0:200], label = "Mean")
plt.ticklabel_format(style='plain', axis='x')
plt.legend()
plt.xlabel("Episodes")
plt.ylabel('Mean')
plt.title('Mean Values of 200,000 Episodes')
plt.figure(figsize=(20, 20))
plt.show()
```
